# Cost of Living Explorer
This notebook loads and visualizes manually collected cost of living data across selected Australian cities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("cost_of_living_template.csv")
df.head()
```

## Clean and Convert Data
Ensure cost column is numeric.

```python
# Drop rows with missing costs or convert to numeric
df['Monthly Cost (AUD)'] = pd.to_numeric(df['Monthly Cost (AUD)'], errors='coerce')
df.dropna(subset=['Monthly Cost (AUD)'], inplace=True)
```

## Total Monthly Cost Per City

```python
total_costs = df.groupby('City')['Monthly Cost (AUD)'].sum().sort_values()
total_costs.plot(kind='barh', title='Total Monthly Cost per City', figsize=(8, 5), color='skyblue')
plt.xlabel('AUD')
plt.tight_layout()
plt.show()
```

## Cost Breakdown by Category

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Monthly Cost (AUD)', y='City', hue='Category')
plt.title('Cost Breakdown by Category')
plt.tight_layout()
plt.show()
```
